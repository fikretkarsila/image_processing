```python
# Threshold Ekleme

img = cv2.imread("data/galata.jpg") # Resmi açtık.
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Resmi gray Çevirdik.

thresh,bfr = cv2.threshold(img,20,255,cv2.THRESH_BINARY) # 20 pixelin üstündekileri 255 renk tonu yaptı.
plt.figure()
plt.imshow(bfr,cmap="gray")
```

```python
img = cv2.imread("data/galata.jpg") # Resmi açtık.
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Resmi gray Çevirdik.

bfr = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,5) # Her pixelin etrafındaki 11 pixele göre bir değer çıkartıyor
plt.figure()  # 11 Sayısını düşürürsek ayrıntı artar
plt.imshow(bfr,cmap="gray")
```

```python
# Blur Ekleme

img = cv2.imread("data/galata.jpg") # Resmi açtık.
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Resmi gray Çevirdik.

bfr = cv2.blur(img,(10,10))
plt.imshow(bfr,cmap="gray")
```

```python
bfr = cv2.GaussianBlur(img,(7,7),0) # 0 X eksenindeki standart sapma değeri.
plt.imshow(bfr,cmap="gray")
```

```python
# Kenar Tespiti

img = cv2.imread("data/galata.jpg") # Resmi açtık.
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Resmi gray Çevirdik.

bfr = cv2.Canny(img,200,255)
plt.imshow(bfr,cmap="gray")
```

```python
bfr = cv2.GaussianBlur(img,(7,7),0) # 0 X eksenindeki standart sapma değeri.
bfr = cv2.Canny(bfr,200,255)
plt.imshow(bfr,cmap="gray")
```

```python
# Histogram

bfr = np.array([[[255,0,0],[0,255,0],[0,0,255],],
                [[0,255,0],[0,0,255],[255,0,0]]],dtype=np.uint8)

bfr.size
bfr.shape

plt.imshow(bfr)
```

```python
color = ("g") # red = r , green = g , blue = b

plt.figure()

for i,k in enumerate(color):
    hist = cv2.calcHist([bfr],channels = [i],mask = None,histSize = [256],ranges = [0,256])
    plt.plot(hist,color=k)
```

```python
# Histogram Eşitleme

bfr = cv2.equalizeHist(img)
plt.imshow(bfr,cmap="gray")
```
